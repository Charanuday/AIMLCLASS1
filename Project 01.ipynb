# Replace the outliers using IQR method

```python
q1=df['Age'].quantile(0.25)
q3=df['Age'].quantile(0.75)
iqr=q3-q1
```

```python
upper_bound=q3+(1.5*iqr)
lower_bound=q1-(1.5*iqr)
```

```python
bat_arr=df['Age']
out=bat_arr[(bat_arr<=lower_bound)|(bat_arr>=upper_bound)]
print(out)
```

```python
ba_arr=df['Age']
lower_cap=np.percentile(ba_arr,0)
upper_cap=np.percentile(ba_arr,65)
```

```python
for i in df['Age']:
    if i <lower_bound:
        df['Age']=df['Age'].replace(i,lower_cap)
    elif i >upper_bound:
        df['Age']=df['Age'].replace(i,upper_cap)
```
